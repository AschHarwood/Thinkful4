```python
import sqlite3 as lite
import pandas as pd
import numpy as np
```

```python
con = lite.connect('social_stats.db')
cur = con.cursor()
```

```python
with con:
    cur.execute('SELECT * FROM education')
    education = cur.fetchall()
    
    cur.execute('SELECT * FROM gdp')
    gdp = cur.fetchall()
```

```python
education_df = pd.DataFrame(education)
```

```python
education_df.head()
```

```python
gdp_df = pd.DataFrame(gdp)
```

```python
gdp_df.head()
```

```python
education_list = list(education_df[1])
gdp_list = list(gdp_df[1])
list(set(education_list) - set(gdp_list))
```
