```python
import sqlite3 as lite
import pandas as pd
```

```python
con = lite.connect('social_stats.db')
cur = con.cursor()
```

```python
with con:
    cur.execute("PRAGMA table_info(gdp)")
    print(cur.fetchall())
```

```python
join_command = 'SELECT country, year, men_years, women_years, total_years, _1999, _2000, _2001, _2002, _2003, _2004, _2005, _2006, _2007, _2008, _2009, _2010 FROM education INNER JOIN gdp ON education.country = gdp.country_name;'


```

```python
with con:
    cur.execute(join_command)
    data = cur.fetchall()
df_combined = pd.DataFrame(data)
```
